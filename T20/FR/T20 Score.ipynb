```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
sns.set_context("poster")
plt.rcParams['axes.prop_cycle'] = plt.cycler(color=plt.cm.Set2.colors)
```

```python
hum = pd.read_csv("blastp_processed_FR_hum.csv")
mus = pd.read_csv("blastp_processed_FR_mus.csv")
nb = pd.read_csv("blastp_processed_FR_nb.csv")
```

```python
plt.figure(figsize=(8,6))
plt.hist(hum['% Identity'],bins=np.arange(50,101,0.5),alpha=0.8,weights = np.ones(hum.shape[0])/hum.shape[0],label="Human")
plt.hist(mus['% Identity'],bins=np.arange(50,101,0.5),alpha=0.8,weights = np.ones(mus.shape[0])/mus.shape[0],label='Murine')
plt.hist(nb['% Identity'],bins=np.arange(50,101,0.5),alpha=0.8,weights = np.ones(nb.shape[0])/nb.shape[0],label='Nanobodies')
plt.legend()
```
